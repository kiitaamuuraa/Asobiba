```
import spacy
from spacy import displacy
```

```
nlp = spacy.load('ja_ginza')

#テキスト読み込み
filepath = "./text/dokujo-tsushin/dokujo-tsushin-6453989.txt"
with open(filepath) as f:
    s = f.read()

#NLPオブジェクトの作成
doc = nlp(s) 
```

```
s
```

```
ent_dict = dict()
for ent in doc.ents:
    ent_dict[ent.text]= ent.label_
```

```
len(ent_dict)
```

```
for i, item in enumerate(ent_dict):
    print(item,":", ent_dict[item])
    if i > 10:
        break
    
```

```
for item in ent_dict:
    if ent_dict[item] == "Person":
        print(item)
```

```
#固有表現抽出の結果の描画

displacy.render(doc, style="ent", jupyter=True)
```
