```
from pycaret.datasets import get_data
boston_data = get_data('boston')
```

```
type(boston_data)
```

```
from pycaret.regression import *
exp1 = setup(boston_data, target = 'medv')
```

```
compare_models()
```

```
cat_b = create_model('catboost')
# 上のセルの1列目を引数にしている
```

```
tuned_cat_b = tune_model(cat_b)
```

```
tuned_cat_b.get_params()
```

```
cat_b._tune_hyperparams
```

```
plot_model(tuned_cat_b)
```

```
cat_b_holdout_pred = predict_model(tuned_cat_b)
```

```
predictions = predict_model(tuned_cat_b, data=boston_data)
```
