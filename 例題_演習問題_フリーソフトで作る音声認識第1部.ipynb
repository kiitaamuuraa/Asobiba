```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

sample =np.array([[3, 2],
          [3, 4],
          [5, 4],
          [5, 6]])
sample = np.random.rand(100, 2)
plt.scatter(sample[:, 0], sample[:, 1])
```

```
mean_sample = sample.mean(axis=0)
std_sample = sample.std(axis=0)
```

```
white_sample = (sample - mean_sample)/std_sample
plt.scatter(white_sample[:, 0], white_sample[:, 1])
```

```
sigma = np.matmul(white_sample, white_sample.T)/(len(white_sample))
```

```
sigma.shape
```

```
import numpy.linalg as LA 
```

```
LA.eig(sigma)[0].shape
```

## 例題4**.3**

```
import scipy.stats as stats
```

```
X = np.array([
              [1,4],
              [2,3],
              [4,3],
              [5,4],
              [2,1],
              [3,2],
              [3,3],
              [4,1]
              ])
Y = np.array([1, 1, 1, 1, 2, 2, 2, 2])
```

```
k = 3
```

```
x = np.array([2.1, 3])
```

```
distance =  np.linalg.norm(X - x, axis=1)
filter = distance.argsort(-1)[:k]
k_nearest_labels = Y[filter]
print(k_nearest_labels)
stats.mode(Y[filter])
```
